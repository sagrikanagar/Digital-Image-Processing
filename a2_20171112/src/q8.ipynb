```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib.colors import hsv_to_rgb
import time
```

```python
def median(im):
    img = im
    w = 2
    h = im.shape[0]
    l = im.shape[1]
    ch = im.shape[2]
    start = time.time()
    for i in range(2,h-2):
        for j in range(2,l-2):
            block = im[i-w:i+w+1,j-w:j+w+1,0]
            m = np.median(block)
            img[i][j][0] = int(m)
            block = im[i-w:i+w+1, j-w:j+w+1,1]
            m = np.median(block)
            img[i][j][1] = int(m)
            block = im[i-w:i+w+1, j-w:j+w+1,2]
            m = np.median(block)
            img[i][j][2] = int(m)
    end = time.time()
    print(end-start)
    return img
```

```python
deg = cv2.imread("../input_data/Degraded.jpg")
clean = cv2.imread("../input_data/Clear.jpg",1)

deg2 = cv2.cvtColor(deg, cv2.COLOR_BGR2HSV)

# print(deg2.shape)
new = median(deg)

```
