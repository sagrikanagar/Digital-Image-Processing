### part a

```python
pic1 = "../input_data/same_hist1.png"
pic2 = "../input_data/same_hist2.png"
img1 = cv2.imread(pic1,0)
img2 = cv2.imread(pic2,0)

histr1 = cv2.calcHist([img1],[0],None,[256],[0,256])
histr2 = cv2.calcHist([img2],[0],None,[256],[0,256])

fig = plt.figure()
plt.subplot(1,2,1)
plt.plot(histr1)
plt.subplot(1,2,2)
plt.plot(histr2)
plt.show()

print("After Matching")

_ = histMatching(pic1,pic2,1)
```
